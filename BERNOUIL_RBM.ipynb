```python
from sklearn.neural_network import BernoulliRBM
import numpy as np

X = np.array([
    [1,1,0],
    [0,1,1],
    [1,0,1],
    [0,0,1]
])

rbm = BernoulliRBM(n_components=2, learning_rate=0.1)

rbm.fit(X)

print("Training Completed")
```
